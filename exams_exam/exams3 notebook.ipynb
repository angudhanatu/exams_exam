# create a new row named total score that adds up the value of all the scores

```python
data['total score'] =data['math score'] + data['writing score'] + data['reading score']
data.head(10)
```

# we similarly create another row named average score to compute the average score per individual

```python
data['average score']= data['total score']/3 #the / gives us the result  of a division as an int(whole number)
data.head()
#the // operator leaves the average score as a float(decimal number)
```

# visualization time

## we are using the seaborn countplot to check gender distribution

       

```python
sns.countplot(data.gender)
plt.title('GENDER DISTRIBUTION')
plt.savefig("gender.png")
plt.show()
```

# we also use seaborn countplot to check racial distribution

```python
sns.countplot(data['race/ethnicity'])
plt.xlabel('RACE')
labels=(data['race/ethnicity'])
group_A_patch = mpatches.Patch(color='blue', label='GROUP A')
group_B_patch = mpatches.Patch(color='red', label='GROUP B')
group_C_patch = mpatches.Patch(color='indigo', label='GROUP C')
group_D_patch = mpatches.Patch(color='orange', label='GROUP D')
group_E_patch = mpatches.Patch(color='green', label='GROUP E')
plt.legend(handles =[group_A_patch, group_B_patch, group_C_patch, group_D_patch, group_E_patch], loc ="upper left")
plt.ylabel('COUNT')
plt.title('TOTAL CLASS DISTRIBUTION BY RACE')
plt.savefig('Racial count.png')
plt.show()
```

```python
plt.figure(figsize =(12,10))
sns.boxplot(data['parental level of education'], data['total score'])
plt.title('MEASURE OF PARENTAL LEVEL OF EDUCATION vs STUDENT PERFORMANCE')
plt.ylabel("AVERAGE SCORE")
plt.xlabel('PARENTAL LEVEL OF EDUCATION')
plt.savefig('parent vs total score.png')
plt.legend()
plt.show()
```

```python
sns.boxplot(data['lunch'], data['total score'])
plt.title('impact of meals on academics')
plt.legend()
plt.savefig('meal vs score.png')
plt.show()
```

```python
plt.figure(figsize =(12,10))
sns.catplot(x='race/ethnicity', y='total score', kind="violin",  data=data)
plt.savefig('boxplot.png', dpi=300)
plt.title('ACADEMIC PERFORMANCE BASE ON RACE/ETHNICITY')
plt.savefig('race vs score.png')

plt.show()
```
